### <a id='toc1_1_5_'></a>[How does the decision boundary change with `k`?](#toc0_)

The bigger is `k`, the more the decision boundary underfit the data and take less account of noisy datapoints.
On the other hand, the smaller is `k` the more the decision boundary **over**fit the data and take is really too precise. Thus less robust to unseen data.

## <a id='toc1_2_'></a>[Task 2: Linear Regression](#toc0_)

**LLS** classifier

${x_i}$: training sample ${i}$

${y}$: label

${f}$: score function ${f = <\omega, x> + b}$

Our goal is to minimize $\mathcal{L}$:
$$
\mathcal{L}=\frac{1}{2} \sum_{n=1}^N\left(f\left(\overline{\mathbf{x}}_i, \theta\right)\right)^2
$$
solved by:
$$
\theta^{\star}=\left(\mathbf{X}^T \mathbf{X}\right)-1 \mathbf{X}^T \mathbf{y}
$$

## <a id='toc1_3_'></a>[Task 3: Softmax Regression & Optimization](#toc0_)


$$
p(C=c \mid \mathbf{x})=\frac{\exp \left(\theta_0^{(c)}+\sum_{j=1}^D \theta_j^{(c)} x_j\right)}{\sum_{k=1}^C \exp \left(\theta_0^{(k)}+\sum_{j=1}^D \theta_j^{(k)} x_j\right)}
$$

With the cross entropy loss to reduce:
$$
\mathcal{L}=- \sum_{i=1}^N \sum_{c=1}^C \log{p(C=c|\mathbf{x}_i) \mathbb{1} [y_i=c]}
$$
