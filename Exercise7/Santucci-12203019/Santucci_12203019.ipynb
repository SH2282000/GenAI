## Task 3: Anomaly Detection using a Variational Autoencder
### 3.1 - Download the Fashion MNIST dataset

We are Download the Fashion MNIST dataset via `torchvision`

```python
from torchvision import datasets, transforms
import torch
from my_vae import loss_vae_wo, reconstruction_loss, VAE
from matplotlib import pyplot as plt
```

```python
device = torch.device("mps" if torch.backends.mps.is_built() else "cpu")
transform = transforms.Compose([transforms.ToTensor()])

train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)

train_dataset_fashion = datasets.FashionMNIST(root='./data', train=True, download=True, transform=transform)
train_loader_fashion = torch.utils.data.DataLoader(train_dataset_fashion, batch_size=64, shuffle=True)
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']



# Extract 10 samples from each class
def extract_1_sample_per_class(dataloader, device):
    sample_per_class = {i: None for i in range(10)}
    for img, label in dataloader.dataset:
        if sample_per_class[label] is None:
            sample_per_class[label] = img.to(device)
        if all(sample is not None for sample in sample_per_class.values()):
            break
    return sample_per_class

sample_per_class_fashion = extract_1_sample_per_class(train_loader_fashion, device)
sample_per_class = extract_1_sample_per_class(train_loader, device)
```

```python
def compute_reconstruction_losses(samples):
    reconstruction_losses = []
    for target, sample in samples.items():
        recon, mu, logvar = model3(sample.unsqueeze(0), torch.tensor([target], device=device))
        with torch.no_grad():
            loss = reconstruction_loss(recon, sample.unsqueeze(0))
            reconstruction_losses.append((loss.item(), recon))

    return reconstruction_losses

reconstruction_losses = compute_reconstruction_losses(sample_per_class)
reconstruction_losses_fashion = compute_reconstruction_losses(sample_per_class_fashion)
```

```python
fig, axes = plt.subplots(10, 4, figsize=(20, 30))
plt.subplots_adjust(hspace=0.5)
fig.suptitle('Anomaly Detection using a Variational Autoencder\nHigher reconstruction loss when given samples from different distribution.', fontsize=20)

for target, sample in sample_per_class.items():
    
    with torch.no_grad():
        axes[target, 0].imshow(sample.squeeze().cpu(), cmap='gray')
        axes[target, 0].set_title(f"Sample from class {target}")
        axes[target, 0].axis('off')
        
        axes[target, 1].imshow(reconstruction_losses[target][1].squeeze().cpu(), cmap='gray')
        axes[target, 1].set_title(f"Reconstructed {target}\nw/ loss: {reconstruction_losses[target][0]:.2f}")
        axes[target, 1].axis('off')
        
        axes[target, 2].imshow(sample_per_class_fashion[target].squeeze().cpu(), cmap='gray')
        axes[target, 2].set_title(f"Sample from class {class_names[target]}")
        axes[target, 2].axis('off')
        
        axes[target, 3].imshow(reconstruction_losses_fashion[target][1].squeeze().cpu(), cmap='gray')
        axes[target, 3].set_title(f"Reconstructed {class_names[target]}\nw/ loss: {reconstruction_losses_fashion[target][0]:.2f}")
        axes[target, 3].axis('off')

plt.show()
```
